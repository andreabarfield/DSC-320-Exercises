```python
# 2a. 
    
# import relevant libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# import data
df = pd.read_csv('customer_spending.csv')

# create histogram
plt.figure(figsize=(8,5))
plt.hist(df['ann_spending'], 
         bins=20,
        edgecolor='black',
        linewidth=1)
plt.xlabel('Annual Spending Amount')
plt.ylabel('Frequency')
plt.title('Annual Customer Spending')
plt.grid(True)
plt.show()
```

```python
# 2b.

# create log transformation using numpy
df['logspending'] = np.log(df['ann_spending'])
```

```python
# 2c.

# create histogram for log transformed dataset
plt.figure(figsize=(8,5))
plt.hist(df['logspending'],
         bins=20,
         edgecolor='black',
         linewidth=1)
plt.xlabel('Log of Annual Spending')
plt.ylabel('Frequency')
plt.title('Log Transformed Customer Spending')
plt.grid(True)
plt.show()
```

```python
# 2d.

# The original data follows more of a left-skewed distribution, 
# where the log transformed data follows more of a bell curve. 
# Log transformed data can reduce the skewness of the 
# original data and make it appear more normal which is better
# for interpretting the data. 
```
