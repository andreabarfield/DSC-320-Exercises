```python
# Problem 1
# Create scatterplot of highway miles per gallon versus weight.

# import libraries
import pandas as pd
import matplotlib.pyplot as plt

# read in car data
df = pd.read_csv("car_data.csv")

# create scatterplot
plt.scatter(df['weight'], df['hwy_mpg'])
plt.xlabel('Weight (lbs)')
plt.ylabel('Highway MPG')
plt.title('Highway MPG versus Vehicle Weight')
plt.show()
```

```python
# Problem 2
# Based on the scatterplot, the relationship between weight and highway miles per 
# gallon is negatively trending. 
```

```python
# Problem 3
# I would expect the slope to be negative because 
# the function is trending negatively. As the weight increases, 
# the highway MPG is decreasing.
```

```python
# Problem 4
# A slope of -0.05 means that for every additional pound of vehicle weight,
# the highway MPG decreases by 0.05 MPG.  
```

```python
# Problem 5
# Add line to graph manually

# import numpy
import numpy as np

# set slope and y-intercept as variables
m = -0.01
b = 60

# set x and y as variables
x = np.array([df['weight'].min(), df['weight'].max()])
y = m*x + b

# create scatterplot again with line
plt.scatter(df['weight'], df['hwy_mpg'])
plt.plot(x, y)
plt.xlabel('Weight (lbs)')
plt.ylabel('Highway MPG')
plt.title('Highway MPG versus Vehicle Weight')
plt.show()
```

```python
# Problem 6
# Find line of best-fit

# import Scikit-learn package
import sklearn
from sklearn.linear_model import LinearRegression

# set variables
X = df[['weight']]
y = df['hwy_mpg']

# create linear regression model
model = LinearRegression()
model.fit(X, y)

# display slope and y-intercept of line of best-fit
print('Slope: ', model.coef_[0])
print('Y-Intercept: ', model.intercept_)

```

```python
# Problem 7
# RMSE of prediction line in problem 5 and actual best-fit line in problem 6.

from sklearn.metrics import root_mean_squared_error

# Predicted value from manual best-fit line
y_pred_man = m * df['weight'] + b

# Calculate RMSE for manual best-fit line and display
rmse_man = root_mean_squared_error(y, y_pred_man)
print(f'Manual Best-Fit Line RMSE: {rmse_man:.4f}')

# Predicted values from regression model
y_pred_model = model.predict(X)

# Calculate RMSE for regression model best-fit line and display
rmse_model = root_mean_squared_error(y, y_pred_model)
print(f'Model Best-Fit Line RMSE: {rmse_model:.4f}')

# The Model RMSE is a smaller value which makes sense because the
# least squares method should minimize the prediction error. 
```

```python
# Problem 8
# Calculate y highway MPG for a vehicle weighing 3200 pounds

Y = model.predict([[3200]])
print(f"Predicted Highway MPG: {Y[0]:.2f}")
```
