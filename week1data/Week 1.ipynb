```python
# Import numpy, pandas, and matplotlib.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt    
```

```python
# Function to calculate RMSE.
def rmse(actual, predicted):

    # Convert input to NumPy arrays.
    actual = np.array(actual)
    predicted = np.array(predicted)

    # Calculate squared difference.
    squared_errors = (predicted - actual) ** 2

    # Calculate mean squared error.
    mse = np.mean(squared_errors)

    # Return square root of MSE.
    return np.sqrt(mse)
```

```python
# Function to calculate MAE.
def mae(actual, predicted):

    # Convert inputs to NumPy arrays.
    actual = np.array(actual)
    predicted = np.array(predicted)

    # Calculate the absolute differences. 
    absolute_errors = np.abs(predicted - actual)

    # Return the average absolute error.
    return np.mean(absolute_errors)
```

```python
# Function to calculate prediction accuracy.
def accuracy(actual, predicted):

    # Convert inputs to NumPy arrays.
    actual = np.array(actual)
    predicted = np.array(predicted)

    # Count number of correct predictions.
    correct_predictions = np.sum(actual == predicted)

    # Calculate accuracy as a percentage.
    accuracy_percent = (correct_predictions / len(actual)) * 100

    # Return the accuracy percentage.
    return accuracy_percent
    
```

```python
# Read housing data.
housing = pd.read_csv('housing_data.csv')

# Read mushroom data.
mushrooms = pd.read_csv('mushroom_data.csv')

# Calculate RMSE.
housing_rmse = rmse(
    housing['sale_price'],
    housing['sale_price_pred']
)

# Calculate MAE.
housing_mae = mae(
    housing['sale_price'],
    housing['sale_price_pred']
)

# Calculate prediction accuracy.
mushroom_accuracy = accuracy(
    mushrooms['actual'],
    mushrooms['predicted']
)

# Display results for problems 1-3.
print('Problem 1')
print(f'RMSE = ${housing_rmse:.2f}')
print('\nProblem 2')
print(f'MAE = ${housing_mae:.2f}')
print('\nProblem 3')
print(f'Accuracy = {mushroom_accuracy:.2f}%')

```

```python
# Define error function.
def error_function(p):
    return(
        0.005 * p**6
        - 0.27 * p**5
        + 5.998 * p**4
        - 69.919  * p**3
        + 449.17 * p**2
        - 1499.7 * p
        + 2028
    )

# Generate x values to estimate the minimum value. 
p = np.linspace(0, 20, 10000)

# Calculate the error for each p value. 
error = error_function(p)

# Calculate the graph.
plt.figure(figsize=(8,5))
plt.plot(p, error)

# Add labels and title.
plt.title('Error Function')
plt.xlabel('Parameter p')
plt.ylabel('Error')
plt.grid(True)

# Diplay graph.
plt.show()

# Find minimum error value and p at that value.
min_error = np.min(error)
best_p = p[np.argmin(error)]

# Display results for problem 4.
print('Problem 4')
print(f'Value of p that minimizes error = {best_p:.4f}')
print(f'Minimum error = {min_error:.4f}')
```
